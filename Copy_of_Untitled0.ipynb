# Классная Работа

```

import matplotlib.pyplot as plt
data = np.loadtxt('http://www.stat.cmu.edu/~larry/all-of-statistics/=data/faithful.dat', skiprows=26)[:, 1:]
plt.scatter(*data.T)
plt.xlabel('Eruption time (min)')
plt.ylabel('Waiting time ti next eruption (min)')
plt.show()
```

```
data = (data - np.mean(data, axis = 0))/ np.std(data, axis=0)
plt.scatter(*data.T)
plt.show()
```

```
centroids = np.random.uniform(-2, 2, 4).reshape((2, 2))
print(centroids)
```

```
dist = [[np.linalg.norm((data - centroids[j])[i]) for j in np.arange(centroids.shape[0])] for i in np.arange(data.shape[0])]
print(np.argmin(dist, axis=1))
```

```
def get_labels(pts, centroids):
  dist = [[np.linalg.norm((pts - centroids[j])[i]) for j in np.arange(centroids.shape[0])] for i in np.arange(pts.shape[0])]
  return np.argmin(dist, axis=1)
labels = get_labels(data, centroids)
print(labels)
```

```
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(*data.T, c=np.where(labels, "#00FFFF", "red"), s=20)
plt.scatter(*centroids.T, c=["red", "#00FFFF"], s=95, marker='*')
plt.title('Predicted Classes')
plt.show()
```
